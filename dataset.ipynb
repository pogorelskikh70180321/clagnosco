```python
from datasets import load_dataset
import pandas as pd
import os

save_dir = r"C:\!project-dataset"
os.makedirs(save_dir, exist_ok=True)

# Loading 'core' dataset
dataset = load_dataset("Salesforce/blip3-kale", name="core", split="train", streaming=True)

batch_size = 50000
batch = []
file_index = 0

for idx, example in enumerate(dataset, 1):
    try:
        batch.append(example)

        if idx % batch_size == 0:
            df = pd.DataFrame(batch)
            file_path = os.path.join(save_dir, f"blip3-kale_part_{file_index:08d}.parquet")
            df.to_parquet(file_path, index=False)
            print(f"Saved {file_path}")
            batch = []
            file_index += 1
    except Exception as e:
        print(f"Error at index {idx}: {e}")
        continue

# Save any remaining examples
if batch:
    df = pd.DataFrame(batch)
    file_path = os.path.join(save_dir, f"blip3-kale_part_{file_index:08d}.parquet")
    df.to_parquet(file_path, index=False)
    print(f"Saved {file_path}")
```

```python
import pandas as pd
import os
import requests
from PIL import Image
from io import BytesIO
from tqdm.notebook import tqdm
import glob
from concurrent.futures import ThreadPoolExecutor, as_completed

input_dir = r"C:\!project-dataset"
output_dir = os.path.join(input_dir, "images")
os.makedirs(output_dir, exist_ok=True)

# Function to detect image format from bytes
def get_image_extension_from_bytes(image_bytes):
    try:
        img = Image.open(BytesIO(image_bytes))
        format_to_ext = {
            "JPEG": "jpg",
            "PNG": "png",
            "WEBP": "webp",
            "BMP": "bmp",
            "GIF": "gif"
        }
        return format_to_ext.get(img.format)
    except Exception as e:
        print(f"Not an image format: {e}")
        return None

# Download function
def download_image(uid, url):
    if not uid or not url:
        return

    existing_files = [f for f in os.listdir(output_dir) if f.startswith(uid + ".")]
    if existing_files:
        return

    try:
        response = requests.get(url, timeout=10)
        if response.status_code == 200:
            ext = get_image_extension_from_bytes(response.content)
            if ext:
                file_path = os.path.join(output_dir, f"{uid}.{ext}")
                with open(file_path, "wb") as f:
                    f.write(response.content)
            else:
                print(f"Unknown format for {uid} from {url}")
        else:
            print(f"Failed {uid} ({response.status_code}): {url}")
    except Exception as e:
        print(f"Error downloading {uid} from {url}: {e}")


# Get list of parquet files
parquet_files = glob.glob(os.path.join(input_dir, "*.parquet"))

for parquet_file in tqdm(parquet_files[20:], desc="Parquet files"):
    print(f"\nProcessing: {os.path.basename(parquet_file)}")
    df = pd.read_parquet(parquet_file)

    # Filter "original_width" and "original_height"
    filtered_rows = df[
        (df["original_width"] > 500) & 
        (df["original_height"] > 500)
    ]

    tasks = [(row["uid"], row["url"]) for _, row in filtered_rows.iterrows()]

    with ThreadPoolExecutor(max_workers=50) as executor:
        futures = [executor.submit(download_image, uid, url) for uid, url in tasks]

        for _ in tqdm(as_completed(futures), total=len(futures), desc="Downloading images", leave=False):
            pass
```

```python
import os
import shutil
import hashlib
from collections import defaultdict
from PIL import Image

def image_hash(filepath):
    """SHA256 hash of pixel data"""
    try:
        with Image.open(filepath) as img:
            img = img.convert("RGB")
            data = img.tobytes()
            return hashlib.sha256(data).hexdigest()
    except Exception as e:
        print(f"Error hashing {filepath}: {e}")
        return None


image_dir = r"C:\!project-dataset\images"
output_dir = r"C:\!project-dataset\duplicates_by_size"
os.makedirs(output_dir, exist_ok=True)

# Step 1: Group by file size
size_dict = defaultdict(list)
for root, dirs, files in os.walk(image_dir):
    for filename in files:
        filepath = os.path.join(root, filename)
        try:
            size = os.path.getsize(filepath)
            size_dict[size].append(filepath)
        except OSError:
            print(f"Couldn't access {filepath}")

# Step 2: Check pixel-perfect matches in the groups
for size, filepaths in size_dict.items():
    hash_dict = defaultdict(list)
    for f in filepaths:
        h = image_hash(f)
        if h:
            hash_dict[h].append(f)

    # Step 3: Copy duplicate images
    for h, duplicates in hash_dict.items():
        if len(duplicates) >= 2:
            size_folder = os.path.join(output_dir, str(size))
            os.makedirs(size_folder, exist_ok=True)
            for f in duplicates:
                filename = os.path.basename(f)

                # Copy to size folder
                dest_path_size = os.path.join(size_folder, filename)
                base, ext = os.path.splitext(filename)
                count = 1
                while os.path.exists(dest_path_size):
                    dest_path_size = os.path.join(size_folder, f"{base}_{count}{ext}")
                    count += 1
                try:
                    shutil.copy2(f, dest_path_size)
                except Exception as e:
                    print(f"Failed to copy to size folder: {f} -> {dest_path_size}: {e}")

                # Copy to common folder
                dest_path_common = os.path.join(output_dir, filename)
                count = 1
                while os.path.exists(dest_path_common):
                    dest_path_common = os.path.join(output_dir, f"{base}_{count}{ext}")
                    count += 1
                try:
                    shutil.copy2(f, dest_path_common)
                except Exception as e:
                    print(f"Failed to copy to common folder: {f} -> {dest_path_common}: {e}")
```

```python
# If image width or height is less than 512 or ratio > 2.5, delete the image
import os
from PIL import Image
from tqdm.notebook import tqdm

image_dir = r"C:\!project-dataset\images"
filenames = os.listdir(image_dir)
error_files = []

for filename in tqdm(filenames, desc="Processing Images"):
    image_path = os.path.join(image_dir, filename)
    try:
        with Image.open(image_path) as img:
            img = img.convert("RGB")
            width, height = img.size
            if min(width, height) < 512 or (max(width, height) / min(width, height) > 2.5):
                os.remove(image_path)
                print(f"Deleted: {image_path}")
    except Exception as e:
        error_files.append(filename)
        print(f"Error processing {image_path}: {e}")
```

```python
# Delete the error files
for filename in error_files:
    image_path = os.path.join(image_dir, filename)
    if os.path.isfile(image_path):
        os.remove(image_path)
        print(f"Deleted: {image_path}")
    else:
        print(f"Not found: {image_path}")
```

RESIZING

```python
import os
from PIL import Image
from tqdm.notebook import tqdm

image_dir = r"C:\!project-dataset\images"
image_resize_dir = r"C:\!project-dataset\images_resize"
os.makedirs(image_resize_dir, exist_ok=True)
image_square_dir = r"C:\!project-dataset\images_square"
os.makedirs(image_square_dir, exist_ok=True)
filenames = os.listdir(image_dir)
new_sizes = dict()
new_sizes_min_max = dict()
resize_size = 192
square_size = 256

for image_file in tqdm(os.listdir(image_dir)):
    image_path = os.path.join(image_dir, image_file)
    resize_path = os.path.join(image_resize_dir, image_file.split(".")[0]+".jpg")
    sqaure_path = os.path.join(image_square_dir, image_file.split(".")[0]+".jpg")

    # Smaller format
    with Image.open(image_path) as img:
        img = img.convert("RGB")
        width, height = img.size

        if width < height:
            new_width = resize_size
            new_height = int((resize_size / width) * height)
        else:
            new_height = resize_size
            new_width = int((resize_size / height) * width)
        new_size = new_width, new_height
        
        if new_size in new_sizes:
            new_sizes[new_size] += 1
        else:
            new_sizes[new_size] = 1
        new_size_min_max = min(new_width, new_height), max(new_width, new_height)
        if new_size_min_max in new_sizes_min_max:
            new_sizes_min_max[new_size_min_max] += 1
        else:
            new_sizes_min_max[new_size_min_max] = 1

        img = img.resize(new_size, Image.LANCZOS)
        img.save(resize_path, "JPEG", quality=100, subsampling=0)
    
    # Square format
    with Image.open(image_path) as img:
        img = img.convert("RGB")
        img = img.resize((square_size, square_size), Image.LANCZOS)
        img.save(sqaure_path, "JPEG", quality=100, subsampling=0)
```

```python
image_dir = r"C:\!project-dataset\images"
caption_dir = r"C:\!project-dataset\captions"
os.makedirs(caption_dir, exist_ok=True)

uid_to_caption = dict(zip(df['uid'].astype(str), df['caption']))

for image_file in tqdm(os.listdir(image_dir), desc="Processing images"):
    if '.' not in image_file:
        continue

    uid = image_file.split('.')[0]

    if uid in uid_to_caption:
        caption = uid_to_caption[uid]
        caption_path = os.path.join(caption_dir, image_file + ".txt")
        with open(caption_path, "w", encoding="utf-8") as f:
            f.write(str(caption))
    else:
        print(f"UID {uid} not found in DataFrame.")
```
